```python
import numpy as np
from PIL import Image
```

```python
group_2 = Image.open('berea8bit.tif')
```

```python
group_2.show()
```

```python
imarray = np.array(group_2)
```

```python
x2 = imarray.reshape((500*500, 1))

from matplotlib import pyplot as plt
plt.hist(x2, density=True, bins=30, range=[0,200]) 
plt.title('Permeability and Porosity')
plt.xlabel('range')
plt.ylabel('freqency')
plt.show()
```

```python
R = imarray
R[R<180] = 0
R[R>=180] = 200
R2 = R.astype(np.bool)
R2 = np.array(R2)
plt.title('Rock grey image')
plt.imshow(imarray, cmap='Greys_r')
```

```python
P = np.size(R2)
framework = np.sum(R2)
porosity = ((P-framework)/P)*100
print('Porosity ',porosity,'%')
```

```python
#another way to find the porosity 
R= imarray
R[R<180]=0
R[R>=180]=255
R2=R.astype(np.bool)
R2=np.array(R2)
r2=np.array(R2)
sout=np.size(r2)
#500*500
areaIn=np.size(r2)
framework=np.sum(r2)
print('Total Volume=',areaIn)
print('Solid Volume=',framework)
```

```python
#Porosity

x=((areaIn-framework)/(areaIn))*100
print("The Porostiy=",x, '%')
```
